### Unsloth

```python
from unsloth import FastLanguageModel
import torch

# 最大序列长度，可自由选择！我们内部自动支持 RoPE Scaling。
max_seq_length = 2048

# 数据类型设置，默认自动检测。Tesla T4 和 V100 使用 Float16，Ampere+ 使用 Bfloat16。
dtype = None

# 是否启用 4bit 量化以减少内存使用。可以设置为 False。
load_in_4bit = True

# 以下是我们支持的 4bit 预量化模型，下载速度提升 4 倍且避免内存不足问题。
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",      # Llama-3.1 模型，15 万亿 tokens，速度提升 2 倍！
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",    # 我们还上传了 405B 的 4bit 模型！
    "unsloth/Mistral-Nemo-Base-2407-bnb-4bit", # 新的 Mistral 12B，速度提升 2 倍！
    "unsloth/Mistral-Nemo-Instruct-2407-bnb-4bit",
    "unsloth/mistral-7b-v0.3-bnb-4bit",        # Mistral v3，速度提升 2 倍！
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",           # Phi-3.5，速度提升 2 倍！
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",            # Gemma，速度提升 2 倍！
]
# 更多模型请访问 https://huggingface.co/unsloth

# 加载模型和分词器
model, tokenizer = FastLanguageModel.from_pretrained(
    # 可选择以下模型：
    # "unsloth/Qwen2.5-0.5B", "unsloth/Qwen2.5-1.5B", "unsloth/Qwen2.5-3B"
    # "unsloth/Qwen2.5-14B",  "unsloth/Qwen2.5-32B",  "unsloth/Qwen2.5-72B"
    # 以及所有 Instruct 版本和数学、编程版本！
    model_name = "unsloth/Qwen2.5-7B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # 如果使用受限模型（如 meta-llama/Llama-2-7b-hf），需要提供 token。
)
```

我们现在添加了 LoRA 适配器，因此我们只需要更新全部参数的 1% 到 10%！

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # 可选择任意大于 0 的数值！推荐值：8、16、32、64、128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # 支持任意值，但设置为 0 是经过优化的
    bias = "none",    # 支持任意值，但设置为 "none" 是经过优化的
    # [新特性] "unsloth" 使用了 30% 更少的显存，可容纳 2 倍更大的批次！
    use_gradient_checkpointing = "unsloth", # 设置为 True 或 "unsloth" 以支持超长上下文
    random_state = 3407,
    use_rslora = False,  # 我们支持 Rank Stabilized LoRA
    loftq_config = None, # 以及 LoftQ
)
```

<a name="Data"></a>
### 数据准备

我们现在使用来自 [yahma](https://huggingface.co/datasets/yahma/alpaca-cleaned) 的 Alpaca 数据集，这是原始 [Alpaca 数据集](https://crfm.stanford.edu/2023/03/13/alpaca.html) 的一个过滤版本，包含 52K 数据。你可以用自己的数据准备代码替换此部分。

**[注意]** 如果只想训练完成任务（忽略用户输入），请参考 TRL 的文档 [这里](https://huggingface.co/docs/trl/sft_trainer#train-on-completions-only)。

**[注意]** 请务必在分词后的输出中添加 **EOS_TOKEN**！否则会导致无限生成问题！

如果你想使用 `llama-3` 模板来处理 ShareGPT 数据集，可以尝试我们的对话式 [笔记本](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Alpaca.ipynb)。

对于文本生成任务（如小说写作），可以尝试这个 [笔记本](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Mistral_(7B)-Text_Completion.ipynb)。


```python
alpaca_prompt = """你是一个信息安全专家，你精通车联网安全、物联网安全、Web安全、无线电安全、移动应用安全等领域，请详细认真的协助我完整各种工作，以下是一个描述任务的指令，与提供进一步上下文的输入配对。请写出一个适当完成请求的回应。

### 指令:
{}

### 输入:
{}

### 回应:
{}"""

EOS_TOKEN = tokenizer.eos_token # 必须添加 EOS_TOKEN
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # 必须添加 EOS_TOKEN，否则生成过程将无限进行！
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("yahma/alpaca-cleaned", split = "train")
dataset = dataset.map(formatting_prompts_func, batched = True,)
```

<a name="Train"></a>
### 训练模型
现在让我们使用 Huggingface TRL 的 `SFTTrainer`！更多文档请参考：[TRL SFT 文档](https://huggingface.co/docs/trl/sft_trainer)。为了加快速度，我们仅进行 60 步训练，但你可以设置 `num_train_epochs=1` 来进行完整训练，并关闭 `max_steps=None`。我们也支持 TRL 的 `DPOTrainer`！


```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False, # 对于短序列，可以使训练速度提高 5 倍。
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1, # 设置此项以进行一次完整的训练。
        max_steps = 60,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # 使用此项可集成到 WandB 等工具
    ),
)
```

```python
# @title 显示当前内存统计信息
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}。最大内存 = {max_memory} GB。")
print(f"{start_gpu_memory} GB 的内存已被保留。")
```

```python
trainer_stats = trainer.train()
```

```python
# @title 显示最终内存和时间统计信息
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"训练耗时 {trainer_stats.metrics['train_runtime']} 秒。")
print(
    f"训练耗时 {round(trainer_stats.metrics['train_runtime']/60, 2)} 分钟。"
)
print(f"峰值保留内存 = {used_memory} GB。")
print(f"训练峰值保留内存 = {used_memory_for_lora} GB。")
print(f"峰值保留内存占最大内存的百分比 = {used_percentage} %。")
print(f"训练峰值保留内存占最大内存的百分比 = {lora_percentage} %。")
```

<a name="Inference"></a>
### 推理
让我们运行模型吧！你可以修改指令和输入——将输出留空即可！

**[新增] 在免费的 Colab 中体验 Llama-3.1 8b Instruct 的2倍速推理 [点击这里](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Unsloth_Studio.ipynb)**


```python
# alpaca_prompt = 从上方复制
FastLanguageModel.for_inference(model)  # 启用原生2倍速推理
inputs = tokenizer(
[
    alpaca_prompt.format(
        "继续斐波那契数列。",  # 指令
        "1, 1, 2, 3, 5, 8",  # 输入
        "",  # 输出 - 留空以生成结果！
    )
], return_tensors = "pt").to("cuda")

outputs = model.generate(**inputs, max_new_tokens = 64, use_cache = True)
tokenizer.batch_decode(outputs)
```

*斜体文本* 你还可以使用 `TextStreamer` 进行连续推理——这样你可以逐个查看生成的 token，而无需等待整个生成过程完成！

```python
# alpaca_prompt = 从上方复制
FastLanguageModel.for_inference(model)  # 启用原生2倍速推理
inputs = tokenizer(
[
    alpaca_prompt.format(
        "继续斐波那契数列。",  # 指令
        "1, 1, 2, 3, 5, 8",  # 输入
        "",  # 输出 - 留空以生成结果！
    )
], return_tensors = "pt").to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer=text_streamer, max_new_tokens=128)
```

```python
if False:
    from unsloth import FastLanguageModel
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name = "lora_model",  # 你用于训练的模型名称
        max_seq_length = max_seq_length,
        dtype = dtype,
        load_in_4bit = load_in_4bit,
    )
    FastLanguageModel.for_inference(model)  # 启用原生2倍速推理

# alpaca_prompt = 你必须从上方复制！

inputs = tokenizer(
[
    alpaca_prompt.format(
        "巴黎著名的高塔是什么？",  # 指令
        "",  # 输入
        "",  # 输出 - 留空以生成结果！
    )
], return_tensors = "pt").to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer=text_streamer, max_new_tokens=128)
```

### 保存为 float16 以用于 VLLM

我们还支持直接保存为 `float16` 格式。选择 `merged_16bit` 保存为 float16，或者选择 `merged_4bit` 保存为 int4。我们还允许使用 `lora` 适配器作为备用方案。使用 `push_to_hub_merged` 可以将模型上传到你的 Hugging Face 账户！你可以访问 https://huggingface.co/settings/tokens 获取你的个人令牌。

```python
# 合并为 16bit
if False: model.save_pretrained_merged("model", tokenizer, save_method = "merged_16bit",)
if False: model.push_to_hub_merged("hf/model", tokenizer, save_method = "merged_16bit", token = "")

# 合并为 4bit
if False: model.save_pretrained_merged("model", tokenizer, save_method = "merged_4bit",)
if False: model.push_to_hub_merged("hf/model", tokenizer, save_method = "merged_4bit", token = "")

# 仅保存 LoRA 适配器
if False: model.save_pretrained_merged("model", tokenizer, save_method = "lora",)
if False: model.push_to_hub_merged("hf/model", tokenizer, save_method = "lora", token = "")
```

### GGUF / llama.cpp 转换

现在我们原生支持保存为 `GGUF` / `llama.cpp` 格式！我们会克隆 `llama.cpp` 并默认保存为 `q8_0`。同时支持所有量化方法，例如 `q4_k_m`。使用 `save_pretrained_gguf` 可以进行本地保存，使用 `push_to_hub_gguf` 可以上传到 Hugging Face。

一些支持的量化方法（完整列表请参考我们的[Wiki 页面](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)）：
* `q8_0` - 快速转换。资源使用较高，但通常可以接受。
* `q4_k_m` - 推荐。对于 attention.wv 和 feed_forward.w2 张量的一半使用 Q6_K，其余部分使用 Q4_K。
* `q5_k_m` - 推荐。对于 attention.wv 和 feed_forward.w2 张量的一半使用 Q6_K，其余部分使用 Q5_K。

[**新功能**] 如果需要微调并自动导出到 Ollama，可以试试我们的 [Ollama notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Ollama.ipynb)。

```python
# 保存为 8bit Q8_0
if False: model.save_pretrained_gguf("model", tokenizer,)
# 请记得访问 https://huggingface.co/settings/tokens 获取一个 token！
# 并将 hf 替换为你的用户名！
if False: model.push_to_hub_gguf("hf/model", tokenizer, token = "")

# 保存为 16bit GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "f16")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "f16", token = "")

# 保存为 q4_k_m GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "q4_k_m")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "q4_k_m", token = "")

# 保存为多个 GGUF 配置选项 - 如果需要多个格式，速度更快！
if False:
    model.push_to_hub_gguf(
        "hf/model", # 将 hf 替换为你的用户名！
        tokenizer,
        quantization_method = ["q4_k_m", "q8_0", "q5_k_m",],
        token = "", # 在 https://huggingface.co/settings/tokens 获取 token
    )
```
