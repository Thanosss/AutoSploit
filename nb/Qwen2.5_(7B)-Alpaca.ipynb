### Unsloth

```python
from unsloth import FastLanguageModel
import torch

# æœ€å¤§åºåˆ—é•¿åº¦ï¼Œå¯è‡ªç”±é€‰æ‹©ï¼æˆ‘ä»¬å†…éƒ¨è‡ªåŠ¨æ”¯æŒ RoPE Scalingã€‚
max_seq_length = 2048

# æ•°æ®ç±»å‹è®¾ç½®ï¼Œé»˜è®¤è‡ªåŠ¨æ£€æµ‹ã€‚Tesla T4 å’Œ V100 ä½¿ç”¨ Float16ï¼ŒAmpere+ ä½¿ç”¨ Bfloat16ã€‚
dtype = None

# æ˜¯å¦å¯ç”¨ 4bit é‡åŒ–ä»¥å‡å°‘å†…å­˜ä½¿ç”¨ã€‚å¯ä»¥è®¾ç½®ä¸º Falseã€‚
load_in_4bit = True

# ä»¥ä¸‹æ˜¯æˆ‘ä»¬æ”¯æŒçš„ 4bit é¢„é‡åŒ–æ¨¡å‹ï¼Œä¸‹è½½é€Ÿåº¦æå‡ 4 å€ä¸”é¿å…å†…å­˜ä¸è¶³é—®é¢˜ã€‚
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",      # Llama-3.1 æ¨¡å‹ï¼Œ15 ä¸‡äº¿ tokensï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",    # æˆ‘ä»¬è¿˜ä¸Šä¼ äº† 405B çš„ 4bit æ¨¡å‹ï¼
    "unsloth/Mistral-Nemo-Base-2407-bnb-4bit", # æ–°çš„ Mistral 12Bï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
    "unsloth/Mistral-Nemo-Instruct-2407-bnb-4bit",
    "unsloth/mistral-7b-v0.3-bnb-4bit",        # Mistral v3ï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",           # Phi-3.5ï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",            # Gemmaï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
]
# æ›´å¤šæ¨¡å‹è¯·è®¿é—® https://huggingface.co/unsloth

# åŠ è½½æ¨¡å‹å’Œåˆ†è¯å™¨
model, tokenizer = FastLanguageModel.from_pretrained(
    # å¯é€‰æ‹©ä»¥ä¸‹æ¨¡å‹ï¼š
    # "unsloth/Qwen2.5-0.5B", "unsloth/Qwen2.5-1.5B", "unsloth/Qwen2.5-3B"
    # "unsloth/Qwen2.5-14B",  "unsloth/Qwen2.5-32B",  "unsloth/Qwen2.5-72B"
    # ä»¥åŠæ‰€æœ‰ Instruct ç‰ˆæœ¬å’Œæ•°å­¦ã€ç¼–ç¨‹ç‰ˆæœ¬ï¼
    model_name = "unsloth/Qwen2.5-7B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # å¦‚æœä½¿ç”¨å—é™æ¨¡å‹ï¼ˆå¦‚ meta-llama/Llama-2-7b-hfï¼‰ï¼Œéœ€è¦æä¾› tokenã€‚
)
```

æˆ‘ä»¬ç°åœ¨æ·»åŠ äº† LoRA é€‚é…å™¨ï¼Œå› æ­¤æˆ‘ä»¬åªéœ€è¦æ›´æ–°å…¨éƒ¨å‚æ•°çš„ 1% åˆ° 10%ï¼

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # å¯é€‰æ‹©ä»»æ„å¤§äº 0 çš„æ•°å€¼ï¼æ¨èå€¼ï¼š8ã€16ã€32ã€64ã€128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # æ”¯æŒä»»æ„å€¼ï¼Œä½†è®¾ç½®ä¸º 0 æ˜¯ç»è¿‡ä¼˜åŒ–çš„
    bias = "none",    # æ”¯æŒä»»æ„å€¼ï¼Œä½†è®¾ç½®ä¸º "none" æ˜¯ç»è¿‡ä¼˜åŒ–çš„
    # [æ–°ç‰¹æ€§] "unsloth" ä½¿ç”¨äº† 30% æ›´å°‘çš„æ˜¾å­˜ï¼Œå¯å®¹çº³ 2 å€æ›´å¤§çš„æ‰¹æ¬¡ï¼
    use_gradient_checkpointing = "unsloth", # è®¾ç½®ä¸º True æˆ– "unsloth" ä»¥æ”¯æŒè¶…é•¿ä¸Šä¸‹æ–‡
    random_state = 3407,
    use_rslora = False,  # æˆ‘ä»¬æ”¯æŒ Rank Stabilized LoRA
    loftq_config = None, # ä»¥åŠ LoftQ
)
```

<a name="Data"></a>
### æ•°æ®å‡†å¤‡

æˆ‘ä»¬ç°åœ¨ä½¿ç”¨æ¥è‡ª [yahma](https://huggingface.co/datasets/yahma/alpaca-cleaned) çš„ Alpaca æ•°æ®é›†ï¼Œè¿™æ˜¯åŸå§‹ [Alpaca æ•°æ®é›†](https://crfm.stanford.edu/2023/03/13/alpaca.html) çš„ä¸€ä¸ªè¿‡æ»¤ç‰ˆæœ¬ï¼ŒåŒ…å« 52K æ•°æ®ã€‚ä½ å¯ä»¥ç”¨è‡ªå·±çš„æ•°æ®å‡†å¤‡ä»£ç æ›¿æ¢æ­¤éƒ¨åˆ†ã€‚

**[æ³¨æ„]** å¦‚æœåªæƒ³è®­ç»ƒå®Œæˆä»»åŠ¡ï¼ˆå¿½ç•¥ç”¨æˆ·è¾“å…¥ï¼‰ï¼Œè¯·å‚è€ƒ TRL çš„æ–‡æ¡£ [è¿™é‡Œ](https://huggingface.co/docs/trl/sft_trainer#train-on-completions-only)ã€‚

**[æ³¨æ„]** è¯·åŠ¡å¿…åœ¨åˆ†è¯åçš„è¾“å‡ºä¸­æ·»åŠ  **EOS_TOKEN**ï¼å¦åˆ™ä¼šå¯¼è‡´æ— é™ç”Ÿæˆé—®é¢˜ï¼

å¦‚æœä½ æƒ³ä½¿ç”¨ `llama-3` æ¨¡æ¿æ¥å¤„ç† ShareGPT æ•°æ®é›†ï¼Œå¯ä»¥å°è¯•æˆ‘ä»¬çš„å¯¹è¯å¼ [ç¬”è®°æœ¬](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Alpaca.ipynb)ã€‚

å¯¹äºæ–‡æœ¬ç”Ÿæˆä»»åŠ¡ï¼ˆå¦‚å°è¯´å†™ä½œï¼‰ï¼Œå¯ä»¥å°è¯•è¿™ä¸ª [ç¬”è®°æœ¬](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Mistral_(7B)-Text_Completion.ipynb)ã€‚


```python
alpaca_prompt = """ä½ æ˜¯ä¸€ä¸ªä¿¡æ¯å®‰å…¨ä¸“å®¶ï¼Œä½ ç²¾é€šè½¦è”ç½‘å®‰å…¨ã€ç‰©è”ç½‘å®‰å…¨ã€Webå®‰å…¨ã€æ— çº¿ç”µå®‰å…¨ã€ç§»åŠ¨åº”ç”¨å®‰å…¨ç­‰é¢†åŸŸï¼Œè¯·è¯¦ç»†è®¤çœŸçš„ååŠ©æˆ‘å®Œæ•´å„ç§å·¥ä½œï¼Œä»¥ä¸‹æ˜¯ä¸€ä¸ªæè¿°ä»»åŠ¡çš„æŒ‡ä»¤ï¼Œä¸æä¾›è¿›ä¸€æ­¥ä¸Šä¸‹æ–‡çš„è¾“å…¥é…å¯¹ã€‚è¯·å†™å‡ºä¸€ä¸ªé€‚å½“å®Œæˆè¯·æ±‚çš„å›åº”ã€‚

### æŒ‡ä»¤:
{}

### è¾“å…¥:
{}

### å›åº”:
{}"""

EOS_TOKEN = tokenizer.eos_token # å¿…é¡»æ·»åŠ  EOS_TOKEN
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # å¿…é¡»æ·»åŠ  EOS_TOKENï¼Œå¦åˆ™ç”Ÿæˆè¿‡ç¨‹å°†æ— é™è¿›è¡Œï¼
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("yahma/alpaca-cleaned", split = "train")
dataset = dataset.map(formatting_prompts_func, batched = True,)
```

<a name="Train"></a>
### è®­ç»ƒæ¨¡å‹
ç°åœ¨è®©æˆ‘ä»¬ä½¿ç”¨ Huggingface TRL çš„ `SFTTrainer`ï¼æ›´å¤šæ–‡æ¡£è¯·å‚è€ƒï¼š[TRL SFT æ–‡æ¡£](https://huggingface.co/docs/trl/sft_trainer)ã€‚ä¸ºäº†åŠ å¿«é€Ÿåº¦ï¼Œæˆ‘ä»¬ä»…è¿›è¡Œ 60 æ­¥è®­ç»ƒï¼Œä½†ä½ å¯ä»¥è®¾ç½® `num_train_epochs=1` æ¥è¿›è¡Œå®Œæ•´è®­ç»ƒï¼Œå¹¶å…³é—­ `max_steps=None`ã€‚æˆ‘ä»¬ä¹Ÿæ”¯æŒ TRL çš„ `DPOTrainer`ï¼


```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False, # å¯¹äºçŸ­åºåˆ—ï¼Œå¯ä»¥ä½¿è®­ç»ƒé€Ÿåº¦æé«˜ 5 å€ã€‚
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1, # è®¾ç½®æ­¤é¡¹ä»¥è¿›è¡Œä¸€æ¬¡å®Œæ•´çš„è®­ç»ƒã€‚
        max_steps = 60,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # ä½¿ç”¨æ­¤é¡¹å¯é›†æˆåˆ° WandB ç­‰å·¥å…·
    ),
)
```

```python
# @title æ˜¾ç¤ºå½“å‰å†…å­˜ç»Ÿè®¡ä¿¡æ¯
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}ã€‚æœ€å¤§å†…å­˜ = {max_memory} GBã€‚")
print(f"{start_gpu_memory} GB çš„å†…å­˜å·²è¢«ä¿ç•™ã€‚")
```

```python
trainer_stats = trainer.train()
```

```python
# @title æ˜¾ç¤ºæœ€ç»ˆå†…å­˜å’Œæ—¶é—´ç»Ÿè®¡ä¿¡æ¯
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"è®­ç»ƒè€—æ—¶ {trainer_stats.metrics['train_runtime']} ç§’ã€‚")
print(
    f"è®­ç»ƒè€—æ—¶ {round(trainer_stats.metrics['train_runtime']/60, 2)} åˆ†é’Ÿã€‚"
)
print(f"å³°å€¼ä¿ç•™å†…å­˜ = {used_memory} GBã€‚")
print(f"è®­ç»ƒå³°å€¼ä¿ç•™å†…å­˜ = {used_memory_for_lora} GBã€‚")
print(f"å³°å€¼ä¿ç•™å†…å­˜å æœ€å¤§å†…å­˜çš„ç™¾åˆ†æ¯” = {used_percentage} %ã€‚")
print(f"è®­ç»ƒå³°å€¼ä¿ç•™å†…å­˜å æœ€å¤§å†…å­˜çš„ç™¾åˆ†æ¯” = {lora_percentage} %ã€‚")
```

### GGUF / llama.cpp è½¬æ¢

* `q5_k_m` - æ¨èã€‚å¯¹äº attention.wv å’Œ feed_forward.w2 å¼ é‡çš„ä¸€åŠä½¿ç”¨ Q6_Kï¼Œå…¶ä½™éƒ¨åˆ†ä½¿ç”¨ Q5_Kã€‚
